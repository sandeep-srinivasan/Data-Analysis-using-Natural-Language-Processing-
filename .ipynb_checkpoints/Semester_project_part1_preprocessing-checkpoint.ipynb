```python
import pandas as pd
import numpy as np
import os
import regex as re
import string
```

```python
class Node():
   def __init__(self):
       # Note that using dictionary for children (as in this implementation) would not allow lexicographic sorting mentioned in the next section (Sorting),
       # because ordinary dictionary would not preserve the order of the keys
       self.children = {}  # mapping from character ==> Node
       self.value = None
   
def find(node, key):
    for char in key:
        if char in node.children:
            node = node.children[char]
        else:
            return None
    return node.value

def update(node, key, difference):
    for char in key:
        if char in node.children:
            node = node.children[char]
    node.value += difference
    
def insert(root, string, value):
    node = root
    index_last_char = None
    for index_char, char in enumerate(string):
        if char in node.children:
            node = node.children[char]
        else:
            index_last_char = index_char
            break

    # append new nodes for the remaining characters, if any
    if index_last_char is not None: 
        for char in string[index_last_char:]:
            node.children[char] = Node()
            node = node.children[char]

    # store value in the terminal node
    node.value = value
```

```python
listtopics=set()
listplaces=set()
cntnotop=0 
cntnoplc=0 
trieTopics = Node()
trieLoc = Node()
WordCount = Node()
topicWordCount = Node()

for i in range(0,22):
    if(i>=10):
        filename = 'reut2-0'+str(i)+'.sgm'
    else:
        filename = 'reut2-00'+str(i)+'.sgm'
    path = 'I:\Users\Sandeep\Desktop\Data Mining Project\Files\\'+filename
    file = open(path, 'rb')
    data = file.read()
#    x = re.findall(r'<REUTERS(.*?)</REUTERS>', data.decode("utf-8"), re.DOTALL, overlapped=True)
    x = re.findall(r'<REUTERS(.*?)</REUTERS>', data.decode("windows-1252"), re.DOTALL, overlapped=True)

    #topicCounter =0
    
    for j in range(0,len(x)):
        # y = re.findall(r'<TOPICS>(.*?)</TOPICS>', x[j].decode("utf-8"), re.DOTALL, overlapped=True)
        y = re.findall(r'<TOPICS>(.*?)</TOPICS>', x[j], re.DOTALL, overlapped=True)
        for k in range(0,len(y)):
            lt = y[k]
            #topics = re.findall(r'<D>(.*?)</D>', lt.decode("utf-8"), re.DOTALL, overlapped=True)
            topics = re.findall(r'<D>(.*?)</D>', lt, re.DOTALL, overlapped=True)
            if(len(topics)==0):
                cntnotop=cntnotop+1
            for l in topics:
                if (find(trieTopics,l) == None):
                    insert(trieTopics, l, 1)
                else:
                    update(trieTopics, l, 1)
                listtopics.add(l)
                #topicCounter += 1
                
    for j in range(0,1):
        z = re.findall(r'<BODY>(.*?)</BODY>', x[j], re.DOTALL, overlapped=True)
        for b,word in enumerate(z):
            body = word.split()
            body = [element.lower() for element in body] ; body
            print(body)
            
            for l in body:
                if (find(WordCount, l) == None):
                    insert(WordCount, l, 1)
                else:
                    update(WordCount, l, 1)
        
    for j in range(0,len(x)):
        # y = re.findall(r'<PLACES>(.*?)</PLACES>', x[j].decode("utf-8"), re.DOTALL, overlapped=True)
        y = re.findall(r'<PLACES>(.*?)</PLACES>', x[j], re.DOTALL, overlapped=True)
        for k in range(0,len(y)):
            lt = y[k]
            places = re.findall(r'<D>(.*?)</D>', lt, re.DOTALL, overlapped=True)
            # places = re.findall(r'<D>(.*?)</D>', lt.decode("utf-8"), re.DOTALL, overlapped=True)
            if(len(places)==0):
                cntnoplc=cntnoplc+1
            for l in places:
                if (find(trieLoc, l) == None):
                    insert(trieLoc, l, 1)
                else:
                    update(trieLoc, l, 1)
                listplaces.add(l)
                #len(listplaces) != len(set(listplaces))
print(listtopics)
print(cntnotop)
print(listplaces)
print(cntnoplc)
```
